```python
df = read_csv('C:/Users/User/Desktop/한국교통안전공단_세종특별자치시 최다 이용 정류장_20221231.csv', encoding='cp949')
df
```

```python
pdf = df.filter(['정류장명','위도','경도'])
pdf
```

```python
icon = folium.Icon(color='red', icon='info-sign')
icon
```

```python
map_osm = folium.Map(location=[36.480058,127.289039], zoom_start=12)

html = "<font color='green' style='white-space: nowrap'><b>%s</b></font>"

# 데이터프레임의 행 수만큼 반복하면서 마커 생성
for i in pdf.index:
    # 행 우선 접근 방식으로 값 추출하기
    name = pdf.loc[i, '정류장명']
    lat = pdf.loc[i, '위도']
    lng = pdf.loc[i, '경도']
    
    # 추출한 정보를 지도에 표시
    popup_html = folium.Popup(html % name, parse_html=False)
    marker = folium.Marker([lat,lng],  icon = folium.Icon(color='red', icon='ok-sign'), popup=popup_html)
    marker2 = folium.CircleMarker([lat,lng], 
                              radius=30,                 # 범위
                              color='red',            # 선 색상
                              fill_color='red')
    marker.add_to(map_osm)
    marker2.add_to(map_osm)
    map_osm.save('교통최다이용지점.html')
map_osm

```

```python
xlsx = read_excel("https://data.hossam.kr/D04/school.xlsx")
xlsx
```

```python
df = xlsx.filter(['학교명', '학교급구분', '소재지도로명주소', '위도', '경도'])
df
```

```python
df2 = df.query("소재지도로명주소.str.contains('세종')")
df2
```

```python
map_osm = folium.Map(location=[37.566651, 126.978428], zoom_start=12)
html = "<font color='green' style='white-space: nowrap'><b>%s</b></font>"

for i in df2.index:
    name = df2.loc[i,'학교명']
    lat = df2.loc[i,'위도']
    lng = df2.loc[i, '경도']

    popup_html = folium.Popup(html % name, parse_html=False)
    marker = folium.Marker([lat,lng], icon = folium.Icon(color='red', icon='flag'), popup=popup_html)
    


    marker.add_to(map_osm)
map_osm
```
