# 시계열 데이터 분석 (정석)

## #01. 작업준비

### 패키지 참조

```python
from pandas import read_excel,  to_datetime
from matplotlib import pyplot as plt
from matplotlib import dates as mdates
from statsmodels.tsa.arima.model import ARIMA
from datetime import timedelta
import seaborn as sb
import sys
sys.path.append("../../")
from helper88 import set_datetime_index, exp_time_data
```

### 그래프 초기화

```python
plt.rcParams["font.family"] = 'AppleGothic' if sys.platform == 'darwin' else 'Malgun Gothic'
plt.rcParams["font.size"] = 12
plt.rcParams["figure.figsize"] = (15, 5)
plt.rcParams["axes.unicode_minus"] = False
```

### 데이터 가져오기

```python
df = read_excel("C:/Users/User/Desktop/세종시_친환경차_보급율_현황(2013_2023).xlsx")
df
```

```python
df.columns
```

```python
df.dtypes
```

```python
df['년도'] = to_datetime(df['년도'], format="%Y%m")
df
```

```python
df = set_datetime_index(df,'년도')
df.head()
```

```python
exp_time_data(data=df, yname="친환경차외(대수)", sd_model='m', max_diff=10)
```

```python
model = ARIMA(df['친환경차외(대수)'], order= (1,1,0), seasonal_order=(1,1,0,3))
fit = model.fit()
print(fit.summary())
```

```python
fv = fit.fittedvalues
fv.head()
```

```python
fc = fit.forecast(365)
fc.head()
```

```python
last = df.index.max()
xmin = last-timedelta(days=365)
xmax = last+timedelta(days=365+10)
ymax = df['친환경차외(대수)'][xmin:xmax].max()
ymin = df['친환경차외(대수)'][xmin:xmax].min()
xmin, xmax, ymax, ymin
```

```python
plt.figure(figsize=(20,10))

# 원본 데이터
sb.lineplot(data=df, x=df.index, y='친환경차외(대수)', label='친환경차외(대수)')

# 원본에 대한 학습결과
sb.lineplot(x=fv.index, y=fv.values, label='FittedValues', linestyle='--')

# 향후 1년간의 예측값
sb.lineplot(x=fc.index, y=fc.values, label='Predict', linestyle='--', color='red')

plt.xlabel('년/월')
plt.ylabel('친환경차외(대수)')
plt.legend()

plt.xlim([xmin, xmax])
plt.ylim([ymin * 0.8, ymax*1.2])

# 그래프의 x축이 날짜로 구성되어 있을 경우 형식 지정
monthyearFmt = mdates.DateFormatter('%y-%m')
plt.gca().xaxis.set_major_formatter(monthyearFmt)

plt.grid()
plt.show()
plt.close()
```

```python
plt.figure(figsize=(15, 5))
sb.boxplot(data=df)
plt.show()
plt.close()
```

## #02. ARIMA 분석 수행

시계열 데이터를 계절 ARIMA 모델에 맞추려고 할 때 첫 번째 목표는 측정항목을 최적화하는 ARIMA(p,d,q)(P,D,Q)s 값을 찾는 것

### 학습 결과에 대한 예측값 생성

```python
start_index = 0
end_index = len(df['친환경차외(대수)'])
pd = fit.predict(start=start_index, end=end_index)
pd
```

### 이후 10 단계의 예측값 생성

```python
fc = fit.forecast(10)
fc
```

### 관측치와 예측치의 비교 결과 시각화

```python
plt.figure(figsize=(20,8))

# 원본
sb.lineplot(x=df.index, y=df['친환경차외(대수)'], label='친환경차외(대수)')

# 원본을 토대로 학습하여 예측한 값
sb.lineplot(x=pd.index, y=pd, label='학습한 예측(대수)', linestyle='--')


sb.lineplot(x=fc.index, y=fc, label='예측(대수)', linestyle='--', color='red')

plt.xlabel('년도')
plt.ylabel('친환경차외(대수)')

# 그래프의 x축이 날짜로 구성되어 있을 경우 형식 지정
monthyearFmt = mdates.DateFormatter('%y.%m')
plt.gca().xaxis.set_major_formatter(monthyearFmt)

plt.grid()
plt.show()
plt.close()
```
